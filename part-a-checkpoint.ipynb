#### Perform data cleaning and text preprocessing

##### Removing stop words, punctuation, and special characters

* Stop Words: These are common words like "the," "a," "an," etc., which don't contribute much meaning to text analysis. You can remove them using libraries like `nltk` or `spaCy.`

* Punctuation and Special Characters: You can remove punctuation (e.g., commas, periods, etc.) and special characters (e.g., @, #) using regex.

##### Tokenization of text (splitting text into words)

* Tokenization refers to splitting the text into individual words or tokens.
* This step is essential for further analysis such as lemmatization or stemming.

##### Lemmatization and Stemming

* Stemming is the process of reducing words to their base or root form (e.g., "running" becomes "run").
* Lemmatization is a more advanced form of reducing words to their root, considering the meaning (e.g., "better" becomes "good").

##### Vectorization using Bag-of-Words and TF-IDF
* Bag-of-Words (BoW): Represents text by counting the frequency of words.
* TF-IDF: Stands for Term Frequency-Inverse Document Frequency, which gives more importance to less frequent words and reduces the weight of common words.

```python
# Initialize NLP preprocessing tools
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove HTML tags if any
    text = re.sub(r'<.*?>', '', text)
    # Remove characters that are not alphabets (punctuation, numbers, etc.)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize by splitting on whitespace
    tokens = text.split()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return " ".join(tokens)
```

```python
# Apply text preprocessing
df["cleaned_review"] = df["review"].apply(preprocess_text)

# Save cleaned data
df.to_csv("cleaned_imdb_data.csv", index=False)
```

```python
# Compute additional text features
df['word_count'] = df['cleaned_review'].apply(lambda x: len(x.split()))
df['char_count'] = df['cleaned_review'].apply(lambda x: len(x))
df['avg_word_length'] = df['cleaned_review'].apply(lambda x: np.mean([len(word) for word in x.split()]) if x.split() else 0)
df['punctuation_count'] = df['review'].apply(lambda x: len(re.findall(r'[^\w\s]', x)))

# Visualize the distribution of these extra features
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['word_count'], bins=50, color='mediumseagreen', kde=True)
plt.title("Word Count Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['avg_word_length'], bins=50, color='salmon', kde=True)
plt.title("Avg. Word Length Distribution")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer

# Ensure the dataset contains the 'cleaned_review' column
if 'cleaned_review' not in df.columns:
    raise KeyError("The column 'cleaned_review' is missing from the dataframe.")

# Initialize and apply TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_text = tfidf_vectorizer.fit_transform(df["cleaned_review"])

# Get TF-IDF feature names and their importance
feature_array = np.array(tfidf_vectorizer.get_feature_names_out())
tfidf_sum = np.asarray(X_text.sum(axis=0)).flatten()  # Sum of TF-IDF scores for each feature

# Get indices of top 20 features
top_n = 20
top_features_idx = np.argsort(tfidf_sum)[-top_n:]
top_features = feature_array[top_features_idx]
top_scores = tfidf_sum[top_features_idx]

# Plot the top TF-IDF features
plt.figure(figsize=(10, 6))
plt.barh(top_features, top_scores, color='royalblue')
plt.xlabel("TF-IDF Score")
plt.ylabel("Top Words")
plt.title("Top 20 Important Words (TF-IDF)")
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

### Model Development

#### Build and train classification models to predict the sentiment of reviews

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg_pred = log_reg.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, log_reg_pred))
print(classification_report(y_test, log_reg_pred))
```

```python
nb = MultinomialNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)

print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_pred))
print(classification_report(y_test, nb_pred))
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```

```python
# Train XGBoost model
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', n_estimators=200, max_depth=6, learning_rate=0.1)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)

# Model Evaluation
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Classification Report:\n", classification_report(y_test, y_pred_xgb))
```

## Model Evaluation

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Define a function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

# Logistic Regression
log_reg_pred = log_reg.predict(X_test)
plot_confusion_matrix(y_test, log_reg_pred, "Logistic Regression")

# Naive Bayes
nb_pred = nb.predict(X_test)
plot_confusion_matrix(y_test, nb_pred, "Naive Bayes")

# Support Vector Machine (SVM)
y_pred_xgb = xgb.predict(X_test)
plot_confusion_matrix(y_test, y_pred_xgb, "XGBoost")

# Random Forest
rf_pred = rf.predict(X_test)
plot_confusion_matrix(y_test, rf_pred, "Random Forest")
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Define function to plot ROC-AUC curves
def plot_roc_auc(model, X_test, y_test, model_name):
    y_prob = model.predict_proba(X_test)[:, 1]  # Get probability scores
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

# Train & Evaluate Models (Replacing SVM with XGBoost)
models = {
    "Logistic Regression": log_reg,
    "Naive Bayes": nb,
    "Random Forest": rf,
    "XGBoost": xgb  # Make sure xgb_model is trained before calling this function
}

plt.figure(figsize=(8, 6))
for model_name, model in models.items():
    plot_roc_auc(model, X_test, y_test, model_name)

plt.plot([0, 1], [0, 1], 'r--')  # Random chance line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC-AUC Curve for IMDb Sentiment Analysis Models")
plt.legend()
plt.show()
```

### Overview
The ROC (Receiver Operating Characteristic) curve is a performance evaluation metric for classification models, plotting the **True Positive Rate (TPR)** against the **False Positive Rate (FPR)** at various threshold levels. The **AUC (Area Under Curve)** score indicates how well a model distinguishes between positive and negative classes.

### Interpretation of the ROC Curve
From the given ROC curve:

**Logistic Regression (AUC = 0.96):**  
  - Achieves the highest AUC score, indicating it is the best-performing model in this analysis.

**Naive Bayes (AUC = 0.93):**  
  - Performs well but slightly underperforms compared to Logistic Regression.

**Random Forest (AUC = 0.93):**

**XGBoost (AUC = 0.93):**
